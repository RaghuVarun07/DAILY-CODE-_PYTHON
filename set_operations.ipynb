SET OPERATIONS 

```python
s4 = {1,2,3,4,5}
s5 = {4,5}
s6 = {6,7,8,9}
```

```python
print(s4)
print(s5)
print(s6)
```

INTERSECTION #It gives common things

```python
s4.intersection(s5)
```

```python
s4 & s6  # & symbol is used for intersection
```

DIFFERENCE #It removes common things and gives remaining elements

```python
s4.difference(s5)
```

Symmetric Difference #It removes the common elements and give the remaining elements fromn both variables


```python
s4.symmetric_difference(s5)
```

```python
# We use the ^ symbol for symmetric difference
s5^s6
```
