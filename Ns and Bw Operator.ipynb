BITWISE OPERATOR

#complement(~)
#and(&)
#or(|)
#xor exclusive or (^)
#left shift(<<)
#right shift(>>)

```python
num1 = 25
~num1
```

```python
# AND (&) (if both are 1,1 then it is 1 otherwise 0)
12 & 13
# binary of 12 is 1100 and binary of 13 is 1101 here and operation is performed between 1100 and 1101
```

```python
# OR (|) (if both are 0,0 then it is 0 otherwise 1)
25|50
```

```python
# XOR (^) (if both are same like(0,0 and 1,1) then it is 0 otherwise 1)
20^10
```

LEFT SHIFT

```python
# In left shift we gain the bits [adding 0 bit to it ex: 12<<1 bin value of 12 is 1100 here at left shift we add 0 to it then it becomes 11000]
12<<1
```

```python
25<<2
```

RIGHT SHIFT

```python
# In right shift we loose the bit 
12>>1
```

```python
25>>2
```
